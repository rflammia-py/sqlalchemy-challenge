# Bonus: Temperature Analysis I

```python
import pandas as pd
from datetime import datetime as dt
```

```python
# "tobs" is "temperature observations"
df = pd.read_csv('hawaii_measurements.csv')
df.head()
```

```python
# Convert the date column format from string to datetime
df["date"] = pd.to_datetime(df["date"])
```

```python
# Set the date column as the DataFrame index
sorted_df = df.set_index("date", drop=False)
sorted_df
```

```python
# Drop the date column
clean_df = sorted_df.drop("date", 1)
clean_df
```

### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Identify the average temperature for June
june_data = clean_df.filter(like = "-06-", axis=0).groupby("date").mean()
june_data
```

```python
# Identify the average temperature for December
dec_data = clean_df.filter(like = "-12-", axis=0).groupby("date").mean()
dec_data
```

```python
# Create collections of temperature data
june = june_data["tobs"]
december = dec_data["tobs"]
```

```python
# Run paired t-test
stats.ttest_ind(june, december)
```

### Analysis

#### With a small p-value and very large T-Test Value, we can say the results are statistically significant. The extremely low p-value suggests that that the results did not happen by random chance; there is a very large and independent difference in temperatures between June and December.
